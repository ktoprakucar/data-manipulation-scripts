```python
import pandas as pd
import numpy as np
import random
import datetime

random.seed(42)
```

```python
df = pd.DataFrame({'date_index': random.sample(range(1, 101), 100),
                   'name': random.choices(['liam', 'noel', 'paul', 'gem'], k=100),
                   'performance': random.choices(['superb', 'good', 'ok', 'bad'], k=100),
                   'salary': random.choices(range(1, 1000), k=100),
                   'working_hours': random.choices(range(20, 50), k=100)})
```

```python
df.sort_values('date_index', inplace=True)
```

```python
df
```

```python
df['last_performance'] = df.groupby('name')['performance'].shift(1)
df['last_1_performance'] = df.groupby('name')['performance'].shift(2)
df['last_2_performance'] = df.groupby('name')['performance'].shift(3)
```

```python
df['moving_working_hours_mean']=df.groupby('name')['working_hours'].shift(1).expanding().mean()
df['last_3_working_hours_mean']=df.groupby('name')['working_hours'].shift(1).rolling(3).mean()
```

```python
df['cumulative_salary']=df.groupby('name')['salary'].shift(1).expanding().sum()
df['last_3_salary']=df.groupby('name')['salary'].shift(1).rolling(3).sum()
```
